```python
import pandas as pd

df = pd.read_csv('insurance.csv')

df
```

```python
import matplotlib.pyplot as plt

# Scatter plot of age vs. charges
plt.scatter(df['age'], df['charges'], alpha=0.5)
plt.title('Scatter Plot of Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python

colors = df['smoker'].map({'yes': 'red', 'no': 'blue'})

# Scatter plot of age vs. charges with different colors for smokers and non-smokers
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['charges'], alpha=0.5, c=colors)
plt.title('Scatter Plot of Age vs Charges (Colored by Smoker Status)')
plt.xlabel('Age')
plt.ylabel('Charges')


#legend
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Smoker'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Non-Smoker')],
           title='Smoker Status')

plt.show()
```

```python
import seaborn as sns


# Set the style of seaborn
sns.set(style="whitegrid")

# Scatter plot of age vs. charges with different colors for smokers and non-smokers using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', hue='smoker', data=df, palette={'yes': 'red', 'no': 'blue'}, alpha=0.7)
plt.title('Scatter Plot of Age vs Charges (Colored by Smoker Status)')
plt.xlabel('Age')
plt.ylabel('Charges')

# Add legend
plt.legend(title='Smoker Status', loc='upper right')

plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

source = ColumnDataSource(df)

# Create a Bokeh figure
p = figure(title='Interactive Scatter Plot with Tooltips', x_axis_label='Age', y_axis_label='Charges')

# Create a scatter plot with different colors for smokers and non-smokers
scatter = p.circle('age', 'charges', source=source, size=10, color='color', legend_field='smoker', alpha=0.7)

# Add tooltips for each point
hover = HoverTool(tooltips=[("Age", "@age"), ("Charges", "@charges")], renderers=[scatter])
p.add_tools(hover)

# Customize the legend
p.legend.title = 'Smoker Status'

# Show the plot
show(p)
```

```python
from plotnine import ggplot, aes, geom_point, scale_color_manual, theme_minimal, labs


# Create a scatter plot with plotnine
scatter_plot = ggplot(df, aes(x='age', y='charges', color='smoker')) + \
    geom_point(alpha=0.7) + \
    scale_color_manual(values={'yes': 'red', 'no': 'blue'}) + \
    theme_minimal() + \
    labs(title='Scatter Plot of Age vs Charges (Colored by Smoker Status)', x='Age', y='Charges')

# Show the plot
scatter_plot.draw()
```
